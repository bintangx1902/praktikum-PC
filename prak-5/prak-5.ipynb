## EDGE FILTERING

```python
def edge_filter(data, filter_size):
    temp = []
    idx = filter_size
    new_img = data.copy()
    row, col = data.shape

    for i in range(row):
        for j in range(col):
            for k in range(i-idx, i+idx+1):
                for m in range(j-idx, j+idx+1):
                    if k > -1 < k < row:
                        if m > -1 and m < col:
                            temp.append(data[k, m])
            temp.remove(data[i, j])
            max_val = max(temp)
            min_val = min(temp)

            if data[i, j] > max_val:
                new_img[i, j] = max_val
            if data[i, j] > max_val:
                new_img[i, j] = max_val

            temp = []

    return new_img.copy()
```
