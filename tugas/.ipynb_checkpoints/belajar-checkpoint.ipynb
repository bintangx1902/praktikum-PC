```python
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
from os.path import dirname, abspath, join
%matplotlib inline
```

```python
path = dirname(dirname(abspath(__name__)))
```

```python
img = cv.imread(join(path, 'image', 'daun-tomat.jpg'))
```

```python
plt.imshow(img)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
hist = cv.calcHist([img], [0], None, [256], [0, 256])
hist_rgb = cv.calcHist([img_rgb], [0], None, [256], [0, 256])
```

```python
def plot_12(instance, histogram):
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))
    axs[0].imshow(instance)
    axs[1].plot(histogram)
```

```python
def plot_12_ravel(instance, color):
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))
    axs[0].imshow(instance)
    axs[1].hist(instance.ravel(), 256, [0, 256], color=color)
```

```python
plot_12(img, hist)
```

```python
plot_12_ravel(img_rgb, 'red')
```

```python
def plot_22(instance, second_instance, histogram, col):
    fig, axs = plt.subplots(2, 2, figsize=(15, 5))
    axs[0, 0].imshow(instance)
    axs[0, 1].plot(histogram, color=col[0])
    
    axs[1, 0].imshow(second_instance)
    axs[1, 1].hist(second_instance.ravel(), 256, [0, 256], color=col[1])
```

```python
def plot_22_ravel(instance, second_instance, col):
    fig, axs = plt.subplots(2, 2, figsize=(15, 5))
    axs[0, 0].imshow(instance)
    axs[0, 1].hist(second_instance.ravel(), 256, [0, 256], color=col[0])
    
    axs[1, 0].imshow(second_instance)
    axs[1, 1].hist(second_instance.ravel(), 256, [0, 256], color=col[1])
```

```python
plot_22(img, img_rgb, hist, ['red', 'pink'])
```

```python
plot_22_ravel(img, img_rgb, ['green', 'purple'])
```

```python
img_gray = cv.imread(join(path, 'image', 'daun-tomat.jpg'), 0)
eq = cv.equalizeHist(img_gray)
img_gray = cv.cvtColor(img_gray, cv.COLOR_BGR2RGB)
```

```python
plot_12_ravel(img_gray, 'blue')
```

```python
height, width, channel = img.shape
```

### mengubah kecerahan

```python
alpha = 20
bright = np.zeros((height, width, channel))

for x in range(height):
    for y in range(width):
        temp = img_rgb[x, y] + alpha
        bright[x, y] = temp

bright = bright.astype(np.uint8)
```

```python
plot_22_ravel(img_rgb, bright, ['pink', 'yellow'])
```
