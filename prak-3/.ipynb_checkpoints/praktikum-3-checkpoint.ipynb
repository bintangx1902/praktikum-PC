```python
import os
from matplotlib import pyplot as plt
import numpy as np
import cv2 as cv
from skimage import data, io, color
% matplotlib inline
```

```python
path = os.path.dirname(os.path.dirname(os.path.abspath(__name__)))
```

```python
path
```

```python
kebo = cv.imread(os.path.join(path, 'image', 'kebo.jpeg'))
```

```python
cv.imshow('real file', kebo)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
kernel = np.array([
    [-1, 0, -1],
    [0, 4, 0],
    [-1, 0, -1],
])

kebo_out = cv.filter2D(kebo, -1, kernel)
```

```python
plt.imshow(kebo)
```

```python
fig, axis = plt.subplots(1, 2, figsize=(15, 11))
ax = axis.ravel()

ax[0].imshow(kebo, cmap='gray')
ax[0].set_title('Data citra asli')
ax[1].imshow(kebo_out, cmap='gray')
ax[1].set_title('data citra output')
```

# HISTOGRAM PLOT

```python
toba = cv.imread(os.path.join(path, 'image', 'danautoba.jpg'))
```

```python
row, column = toba.shape[:2]
```

```python
hist = cv.calcHist([toba], [0], None, [256], [0, 256])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(toba)
axs[1].plot(hist)
```

```python
img = cv.cvtColor(toba, cv.COLOR_BGR2RGB)
```

```python
hist = cv.calcHist([img], [0], None, [256], [0, 256])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(img)
axs[1].plot(hist, color='g')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0, 0].imshow(img)
axs[0, 1].plot(hist, color='#8B008B')

axs[1, 0].imshow(img)
axs[1, 1].hist(img.ravel(), 256, [0, 256], color='#FF1493')
plt.show()
```

```python
gray_toba = cv.imread(os.path.join(path, 'image', 'danautoba.jpg'), 0)
hist2 = cv.equalizeHist(gray_toba)

fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(cv.cvtColor(gray_toba, cv.COLOR_BGR2RGB))
axs[1].hist(gray_toba.ravel(), 256, [0, 256], color='g')
# axs[1].plot(hist2, color='g')
```

```python
beta = 35
bright = np.zeros((row, column, 3))

for x in range(row):
    for y in range(column):
        gyx = img[x, y] + beta
        bright[x, y] = gyx

bright = bright.astype(np.uint8)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0, 0].imshow(img)
axs[0, 1].hist(gray_toba.ravel(), 256, [0, 256], color='red')

axs[1, 0].imshow(bright)
axs[1, 1].hist(bright.ravel(), 256, [0, 256], color='#67f2d1')
plt.show()
```

```python
beta = 40
alpha = 1.2
result = np.zeros((row, column, 3))

for x in range(row):
    for y in range(column):
        gyx = alpha * img[x, y] + beta
        result[x,y] = gyx

result = result.astype(np.uint8)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0, 0].imshow(img)
axs[0, 1].hist(gray_toba.ravel(), 256, [0, 256], color='red')

axs[1, 0].imshow(result)
axs[1, 1].hist(result.ravel(), 256, [0, 256], color='#67f2d1')
plt.show()
```
