```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
from os import path
```

```python
path_ = path.dirname(path.dirname(path.abspath(__name__)))
```

```python
path
```

```python
img = cv.imread(path.join(path_, 'image', 'parking_lot.jpg'))
```

```python
plt.imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
edge = cv.Canny(img, 100, 150)
```

```python
plt.imshow(edge)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
ax = axs.ravel()

ax[0].imshow(img, cmap='gray')
ax[0].set_title('gambar asli')

ax[1].imshow(edge, cmap='gray')
ax[1].set_title('gambar tepi')
```

```python
lines = cv.HoughLinesP(edge, 1, np.pi/180, 30, maxLineGap=250)
img_line = img.copy()
```

```python
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv.line(img_line, (x1, y1), (x2,y2), (0, 0, 255), 1)
```

```python
plt.imshow(img_line)
```

```python
gray1 = cv.cvtColor(img_line, cv.COLOR_BGR2RGB)

fig, axs = plt.subplots(1, 2, figsize=(15, 5))
ax = axs.ravel()

ax[0].imshow(gray, cmap='gray')
ax[0].set_title('gambar asli')

ax[1].imshow(gray1, cmap='gray')
ax[1].set_title('gambar garis parkiran')
```
