# PENGAMBANGAN DWI ARAS DENGAN PERIKIRAAN NILAI AMBANG

```python
thresh = 128
h, w = leaf.shape
leaf_result = leaf.copy()
for x in range(h):
    for y in range(w):
        leaf_result[x, y] = 0 if leaf[x, y] < thresh else 255
```

```python
plt.imshow(leaf_result)
```

```python
cv.imshow('gambar hasil', leaf_result)
cv.waitKey(0)
cv.destroyAllWindows()
```

## MENENTUKAN NILAI AMBANG SUATU GAMBAR

```python
def titerative(image):
    h, w = image.shape
    t0 = 127
    while True:
        avg_left, avg_right = 0, 0
        sum_left, sum_right = 0, 0
        for x in range(h):
            for y in range(w):
                if image[x, y] <= 127:
                    avg_left += image[x, y]
                    sum_left += 1
                else:
                    avg_right += image[x, y]
                    sum_right += 1
        avg_left /= sum_left
        avg_right /= sum_right
        t1 = (avg_right + avg_left) / 2
        if (t0 - t1) < 1:
            break
        t0 = t1
    return round(t1)
```

```python
new_thresh = titerative(leaf)
```
