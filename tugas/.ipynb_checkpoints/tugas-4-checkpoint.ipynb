# HISTOGRAM

```python
row, col = tomato.shape[:2]
```

```python
hist = cv2.calcHist([tomato], [0], None, [256], [0, 256])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(tomato)
axs[1].plot(hist)
```

```python
img = cv2.cvtColor(tomato, cv2.COLOR_BGR2RGB)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(img)
axs[1].plot(hist, color='g')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0, 0].imshow(img)
axs[0, 1].plot(hist, color='#8B008B')

axs[1, 0].imshow(img)
axs[1, 1].hist(img.ravel(), 256, [0, 256], color='#FF1493')
plt.show()
```

```python
hist2 = cv2.equalizeHist(tomato_gray)

fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(cv2.cvtColor(tomato_gray, cv2.COLOR_BGR2RGB))
axs[1].hist(tomato_gray.ravel(), 256, [0, 256], color='g')
```

### DI CERAHKAN

```python
beta = 15
bright = np.zeros((row, col, 3))

for x in range(row):
    for y in range(col):
        temp = img[x, y] + beta
        bright[x, y] = temp

bright = bright.astype(np.uint8)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0, 0].imshow(img)
axs[0, 1].hist(img.ravel(), 256, [0, 256], color='red')

axs[1, 0].imshow(bright)
axs[1, 1].hist(bright.ravel(), 256, [0, 256], color='#67f2d1')
plt.show()
```

## MEREGANGKAN KECERAHAN

```python
beta = 20
alpha = .9
result = np.zeros((row, col, 3))

for x in range(row):
    for y in range(col):
        gyx = alpha * img[x, y] + beta
        result[x,y] = gyx

result = result.astype(np.uint8)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0, 0].imshow(img)
axs[0, 1].hist(img.ravel(), 256, [0, 256], color='red')

axs[1, 0].imshow(result)
axs[1, 1].hist(result.ravel(), 256, [0, 256], color='#67f2d1')
plt.show()
```

## MENAMBAHKAN KONTRAS

```python
alpha = .8
contrast = np.zeros((row, col, 3))

for x in range(row):
    for y in range(col):
        gyx = alpha * img[x, y]
        contrast[x,y] = gyx

contrast = contrast.astype(np.uint8)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0, 0].imshow(img)
axs[0, 1].hist(img.ravel(), 256, [0, 256], color='red')

axs[1, 0].imshow(contrast)
axs[1, 1].hist(contrast.ravel(), 256, [0, 256], color='#67f2d1')
plt.show()
```

## MENGKOMBINASIKAN KONTRAS DAN KECERAHAN

```python
combined = bright + contrast
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0, 0].imshow(img)
axs[0, 1].hist(img.ravel(), 256, [0, 256], color='red')

axs[1, 0].imshow(combined)
axs[1, 1].hist(combined.ravel(), 256, [0, 256], color='#67f2d1')
plt.show()
```
