```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
import os
%matplotlib inline
```

```python
path = os.path.dirname(os.path.dirname(os.path.abspath(__name__)))
```

## menampilkan gambar

```python
citra1 = imread(fname=os.path.join(path, 'image', 'noisy.jfif'))
citra2 = imread(fname=os.path.join(path, 'image', 'dog.jfif'))

citra1 = cv.cvtColor(citra1, cv.COLOR_BGR2GRAY)
citra2 = cv.cvtColor(citra2, cv.COLOR_BGR2GRAY)
```

```python
print(f"{citra1.shape = }")
print(f"{citra2.shape = }")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs = axs.ravel()

axs[0].imshow(citra1, cmap='gray')
axs[1].imshow(citra2, cmap='gray')

axs[0].set_title('Citra 1')
axs[1].set_title('Citra 2')
```

## copy gambar

```python
img1 = citra1.copy().astype(float)
img2 = citra2.copy().astype(float)

m1, n1 = img1.shape
m2, n2 = img2.shape
output1 = np.empty([m1, n1])
output2 = np.empty([m2, n2])

print(f"{img1.shape = }")
print(f"{output1.shape = }")

print(f"{m1 = }", f"{n1 = }", sep='\n')
print()

print(f"{img1.shape = }")
print(f"{output1.shape = }")

print(f"{m1 = }", f"{n1 = }", sep='\n')
```

```python
for row in range(0, m1-1):
    for col in range(0, n1-1):
        a1, b1 = row, col
        result = img1[a1-1, b1-1] + img1[a1-1, b1] + img1[a1-1, b1-1] + \
                 img1[a1, b1-1] + img1[a1, b1] + img1[a1, b1+1] + \
                 img1[a1+1, b1-1] + img1[a1+1, b1] + img1[a1+1, b1+1]

        output1[a1, b1] = 1/9 * result
```

```python
for row in range(0, m2-1):
    for col in range(0, n2-1):
        a1, b1 = row, col
        result = img2[a1-1, b1-1] + img2[a1-1, b1] + img2[a1-1, b1-1] + \
                 img2[a1, b1-1] + img2[a1, b1] + img2[a1, b1+1] + \
                 img2[a1+1, b1-1] + img2[a1+1, b1] + img2[a1+1, b1+1]

        output2[a1, b1] = 1/9 * result
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.ravel()

axs[0].imshow(citra1, cmap='gray')
axs[0].set_title('input citra 1')

axs[1].imshow(citra2, cmap='gray')
axs[1].set_title('input citra 2')

axs[2].imshow(img1, cmap='gray')
axs[2].set_title('output citra 1')

axs[3].imshow(img2, cmap='gray')
axs[3].set_title('output citra 2')
```
